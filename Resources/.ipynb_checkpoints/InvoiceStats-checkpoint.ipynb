```python
#import dependencies
import pandas as pd
```

```python
#create csv path
stats = "InvoiceStats.csv"

# Read the CSV into a Pandas DataFrame
inv_stats = pd.read_csv(stats)

inv_stats.head()
```

```python
#Reorganize columns and drop unnecessary columns

inv_stats = inv_stats[["Servicing Dealer", "Billing Group", "Dealer Invoice", "Invoice Status", "Approved Amount", "Invoice Create Date", "Invoice Issue?"]]

inv_stats.head()
```

```python
#rename columns

inv_stats = inv_stats.rename(columns={"Servicing Dealer":"Branch", "Billing Group":"Customer", "Dealer Invoice":"Inv_Number", "Invoice Status":"Inv_Status", "Approved Amount":"App_Amt", "Invoice Create Date":"Inv_Date", "Invoice Issue?":"Inv_Issue"})


inv_stats.head()
```

```python
inv_stats.dtypes
```

```python
#update Inv Date to datetime
inv_stats["Inv_Date"] = pd.to_datetime(inv_stats["Inv_Date"])

inv_stats.dtypes
```

```python
#rename branch number to abbreviation
inv_stats = inv_stats.replace({60789:"DEN", 60790:"DEN", 60791:"SLC", 60792:"ABQ"})

inv_stats.head()
```

```python
#view customer values so abbreviations can be updated to customer name
inv_stats["Customer"].value_counts()
```

```python
#updated customer abbreviations
inv_stats = inv_stats.replace({" AZ":"Amazon", " HD":"Home Depot", " WM":"Walmart", " ML":"Medline", " UA":"US Autoforce",
                                                " KH":"KeHE", " ITL":"HD Pro", " HS":"HD Supply", " XPO":"GXO Logistics", " SC":"Sam's Club", " ARM": "Aramsco",
                                                " KN": "Kuehne + Nagel", " WW":"Grainger", " TCI":"TBC International", " XL":"DHL", " BZ":"Bunzl",
                                                " TSI":"Sealy Mattress", " MR":"Lineage Logistics", " MFD":"Mattress Firm", " OD":"Office Depot", " CO":"Costco",
                                                " AR":"Americold", " HF":"Harbor Freight", " HDD":"HD Pro", " PCS":"Petco", " BKR":"Baker Dist", " COL":"Composites One", " WAY":"Wayfair", 
                                                " GS":"Gensco", " BL":"Big Lots", " PTN":"Peloton", " LME":"XPO Last Mile", " BBB": "Bed Bath Beyond"})

inv_stats.head()
```

```python
#verify all customer names updated
inv_stats["Customer"].value_counts()
```

```python
#create new column for month 
inv_stats["Month"] = inv_stats["Inv_Date"].dt.month
inv_stats.head()
```

```python
#replace Month number with name

inv_stats["Month"] = inv_stats["Month"].replace({1:"January", 2:"February", 3:"March", 4:"April", 5:"May", 6:"June", 
                                                         7:"July", 8:"August", 9:"September", 10:"October", 11:"November", 
                                                         12:"December"})
```

```python
#verify all values have updated
inv_stats["Month"].value_counts()
```

```python
#create new dataframe with summary of total number of invoices, invoices with and without errors and percentage

#branch total invoices
branch_total = inv_stats["Branch"].value_counts()

#branch inv count with errors
#filter main DF to only show "Yes" in Inv_Issue column
stats = inv_stats[inv_stats.Inv_Issue == "Yes"]
#count of invoices with issues per branch
issue = stats.groupby(["Branch"]).count()["Inv_Issue"]

#subtract inv issues from total to calculate no issues
no_issue = branch_total - issue

#calculate percentage with error
percent = issue/branch_total *100
```

```python
#new DF
stat_summary = pd.DataFrame({"Total_Inv": branch_total, "Issue": issue, "No_Issue":no_issue, "Issue%":percent})

stat_summary
```

```python
#caculate totals for each column

summary_total = stat_summary["Total_Inv"].sum()
issue_total = stat_summary["Issue"].sum()
no_issue_total = stat_summary["No_Issue"].sum()
percent_total = issue_total/summary_total *100
```
