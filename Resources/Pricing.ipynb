```python
#import dependencies
import pandas as pd
```

```python
#create csv path
pricing = "Pricing.csv"

# Read the CSV into a Pandas DataFrame
invpricing_df = pd.read_csv(pricing)

invpricing_df.head()
```

```python
#Reorgainze columns
invpricing_df = invpricing_df[["Branch", "Dept Impacted", "Customer Name", "Invoice", "Work Order", " QW Total ", " SAP Total ", " Discrepancy ", "Cause", "Date"]]

invpricing_df.head()
```

```python
#rename columns
invpricing_df = invpricing_df.rename(columns={"Customer Name":"Customer", "Dept Impacted":"Dept_Impacted", "Invoice":"Inv_Number", "Work Order":"WO", 
                                              " QW Total ":"QW_Total", " SAP Total ":"SAP_Total", " Discrepancy ":"Discrepancy", "Date":"Inv_Date"})

invpricing_df.head()
```

```python
#Combine branch and dept 

invpricing_df['Branch_Dept'] = invpricing_df.Branch.str.cat(invpricing_df.Dept_Impacted,sep=" ")
invpricing_df
```

```python
#create new DF with branch_dept combined column
invpricing = invpricing_df[["Branch_Dept", "Customer", "Inv_Number", "WO", "QW_Total", "SAP_Total", "Discrepancy", "Cause", "Inv_Date"]]

invpricing.head()
```

```python
#check data types to adjust as needed
invpricing.dtypes
```

```python
#update SAP Total column to numeric values and Date column to datetime

invpricing["Inv_Date"]=pd.to_datetime(invpricing["Inv_Date"])

#verify data types updated
invpricing.dtypes
```

```python
invpricing.head()
```

```python
#create new column for month 
invpricing["Month"] = invpricing["Inv_Date"].dt.month
invpricing.head()
```

```python
#replace Month number with name

invpricing["Month"] = invpricing["Month"].replace({1:"January", 2:"February", 3:"March", 4:"April", 5:"May", 6:"June", 
                                                         7:"July", 8:"August", 9:"September", 10:"October", 11:"November", 
                                                         12:"December"})
```

```python
#verify all values have updated
invpricing["Month"].value_counts()
```

```python
#view all cause categories
invpricing["Cause"].value_counts()
```

```python
#combine Causes into main categories
invpricing["Cause"] = invpricing["Cause"].replace({"Part Pricing - Raymond": "Part_Pricing", "Part Pricing - TVH/Durasource":"Part_Pricing",
                                                  "Part Pricing - Misc":"Part_Pricing", "Tax":"Tax_Enviro", "Environmental Fee":"Tax_Enviro", "Over NTE":"Over_NTE", 
                                                   "PO funds":"Over_NTE", "PO Funds":"Over_NTE", "Insufficient PO Funds":"Over_NTE", "Incorrect Labor Rate":"Incorrect_Rate",
                                                  "Discount Labor Line":"Incorrect_Rate", "Incorrect PM Rate":"Incorrect_Rate", "Incorrect Part Pricing":
                                                  "Incorrect_Rate", "Incorrect Labor Rate/Amt":"Incorrect_Rate" , "Productive Rounding":"Incorrect_Rate",
                                                  "Entry Error - Labor Quantity":"Entry_Error", "Entry Error - Freight": "Entry_Error", 
                                                    "Entry Error - Travel Quantity":"Entry_Error", "Entry Error - Misc Charge":"Entry_Error", 
                                                   "Entry Error - Part Pricing":"Entry_Error", "Entry Error - Travel Rate":"Entry_Error", "Entry Error - Part Quantity":
                                                  "Entry_Error", "Entry Error - SM Pricing":"Entry_Error", "PDN":"Misc_Charge", "Tire Press":"Misc_Charge", 
                                                   "Tire Disposal":"Misc_Charge", "Entry Error - Tire Press":"Misc_Charge"}) 


invpricing["Cause"].value_counts()
```

```python
#create new DF with DEN Service invoices
den_df = invpricing[invpricing.Branch_Dept == "DEN Service"]

den_df.head()
```

```python
#calculate monthly discrepancy for Denver Service
den_monthly = den_df.groupby(["Month"]).sum("Discrepancy")

den_monthly = den_monthly["Discrepancy"]

den_monthly 
```

```python
#Calculate $$ per cause category
den_categorical = den_df.groupby(["Cause"]).sum("Discrepancy")

den_categorical = den_categorical["Discrepancy"]

den_categorical
```

```python
#create new DF with month and cause $$ amounts
den_service = pd.DataFrame({"January":[den_monthly.January], "February":[den_monthly.February], "March":[den_monthly.March],
                           "April":[den_monthly.April], "May":[den_monthly.May], "June":[den_monthly.June], "July":[den_monthly.July],
                           "August":[den_monthly.August], "September":[den_monthly.September], "October":[den_monthly.October], 
                           "November":[den_monthly.November], "December":[den_monthly.December], "Part_Pricing":[den_categorical.Part_Pricing],
                           "Over_NTE":[den_categorical.Over_NTE], "Consumables":[den_categorical.Consumables], 
                            "Entry_Error":[den_categorical.Entry_Error], "Incorrect_Rate":[den_categorical.Incorrect_Rate], "Misc_Charge":[den_categorical.Misc_Charge],
                           "Tax_Enviro":[den_categorical.Tax_Enviro]})



#rename index to branch/dept
denserv_summary=den_service.rename({0: "Den_Service"})

denserv_summary
```
