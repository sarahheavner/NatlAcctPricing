```python
#import dependencies
import pandas as pd
```

```python
#create csv path
pricing = "Pricing.csv"

# Read the CSV into a Pandas DataFrame
inv_pricing = pd.read_csv(pricing)

inv_pricing.head()
```

```python
#Reorgainze columns
invpricing_df = inv_pricing[["Branch", "Dept Impacted", "Customer Name", "Invoice", "Work Order", " QW Total ", " SAP Total ", " Discrepancy ", "Cause", "Date"]]

invpricing_df.head()
```

```python
#rename columns
invpricing_df = invpricing_df.rename(columns={"Customer Name":"Customer", "Invoice":"Inv Number", "Work Order":"WO", 
                                              " QW Total ":"QW Total", " SAP Total ":"SAP Total", " Discrepancy ":"Discrepancy"})

invpricing_df.head()
```

```python
#check data types to adjust as needed
invpricing_df.dtypes
```

```python
#replace "-" value with 0 in order to convert to numeric
invpricing_df['SAP Total'] = invpricing_df['SAP Total'].replace({ "-": "0.00"})
```

```python
#update QW Total, SAP Total, Discrepancy to numeric values
invpricing_df["SAP Total"] = invpricing_df["SAP Total"].apply(pd.to_numeric)

#verify data types updates
invpricing_df.dtypes
```
