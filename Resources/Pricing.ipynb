```python
#import dependencies
import pandas as pd
```

```python
#create csv path
pricing = "Pricing.csv"

# Read the CSV into a Pandas DataFrame
invpricing_df = pd.read_csv(pricing)

invpricing_df.head()
```

```python
#Reorgainze columns
invpricing_df = invpricing_df[["Branch", "Dept Impacted", "Customer Name", "Invoice", "Work Order", " QW Total ", " SAP Total ", " Discrepancy ", "Cause", "Date"]]

invpricing_df.head()
```

```python
#rename columns
invpricing_df = invpricing_df.rename(columns={"Customer Name":"Customer", "Dept Impacted":"Dept_Impacted", "Invoice":"Inv_Number", "Work Order":"WO", 
                                              " QW Total ":"QW_Total", " SAP Total ":"SAP_Total", " Discrepancy ":"Discrepancy", "Date":"Inv_Date"})

invpricing_df.head()
```

```python
#Combine branch and dept 

invpricing_df['Branch_Dept'] = invpricing_df.Branch.str.cat(invpricing_df.Dept_Impacted,sep=" ")
invpricing_df
```

```python
#create new DF with branch_dept combined column
invpricing = invpricing_df[["Branch_Dept", "Customer", "Inv_Number", "WO", "QW_Total", "SAP_Total", "Discrepancy", "Cause", "Inv_Date"]]

invpricing.head()
```

```python
#check data types to adjust as needed
invpricing.dtypes
```

```python
#update SAP Total column to numeric values and Date column to datetime

invpricing["Inv_Date"]=pd.to_datetime(invpricing["Inv_Date"])

#verify data types updated
invpricing.dtypes
```

```python
invpricing.head()
```

```python
#create new column for month 
invpricing["Month"] = invpricing["Inv_Date"].dt.month
invpricing.head()
```

```python
#replace Month number with name

invpricing["Month"] = invpricing["Month"].replace({1:"January", 2:"February", 3:"March", 4:"April", 5:"May", 6:"June", 
                                                         7:"July", 8:"August", 9:"September", 10:"October", 11:"November", 
                                                         12:"December"})
```

```python
#verify all values have updated
invpricing["Month"].value_counts()
```

```python
#create new DF with DEN invoices
den_df = invpricing[invpricing.Branch_Dept == "DEN Service"]

den_df.head()
```

```python
den_monthly = den_df.groupby(["Month"]).sum("Discrepancy")

den_monthly
```
