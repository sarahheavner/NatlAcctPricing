```python
#import dependencies
import pandas as pd
```

```python
#create csv path
pricing = "Pricing.csv"

# Read the CSV into a Pandas DataFrame
invpricing_df = pd.read_csv(pricing)

invpricing_df.head()
```

```python
#Reorgainze columns and drop QW Total and SAP Total(unnecessary columns)
invpricing_df = invpricing_df[["Branch", "Dept Impacted", "Customer Name", "Invoice", "Work Order", " Discrepancy ", 
                               "Cause", "Date"]]

invpricing_df.head()
```

```python
#rename columns
invpricing_df = invpricing_df.rename(columns={"Customer Name":"Customer", "Dept Impacted":"Dept_Impacted", "Invoice":"Inv_Number", "Work Order":"WO", 
                                               " Discrepancy ":"Discrepancy", "Date":"Inv_Date"})

invpricing_df.head()
```

```python
#check data types to adjust as needed
invpricing_df.dtypes
```

```python
#update Date column to datetime

invpricing_df["Inv_Date"] = pd.to_datetime(invpricing_df["Inv_Date"])

invpricing_df.head()
```

```python
#create new column for month 
invpricing_df["Month"] = invpricing_df["Inv_Date"].dt.month
invpricing_df.head()
```

```python
#replace Month number with name

invpricing_df["Month"] = invpricing_df["Month"].replace({1:"January", 2:"February", 3:"March", 4:"April", 5:"May", 6:"June", 
                                                         7:"July", 8:"August", 9:"September", 10:"October", 11:"November", 
                                                         12:"December"})

invpricing_df
```

```python
#Combine branch and dept 

invpricing_df['Branch_Dept'] = invpricing_df.Branch.str.cat(invpricing_df.Dept_Impacted,sep="_")
invpricing_df
```

```python
#create new DF with branch_dept combined column, drop Branch and Department columns (unnecessary/duplicate data)
invpricing = invpricing_df[["Branch_Dept", "Customer", "Inv_Number", "WO", "Discrepancy", "Cause", "Inv_Date", "Month"]]

invpricing.head()
```

```python
invpricing["Branch_Dept"].value_counts()
```

```python
#update Mountain Power and Denver Tire values 
invpricing = invpricing.replace({"ABQ_Mountain Power":"MountainPower", "DEN_Mountain Power":"MountainPower", "DEN_Tire Truck":"DEN_Tire"})

invpricing
```

```python
#view all cause categories
invpricing["Cause"].value_counts()
```

```python
#combine Causes into similar categories
invpricing = invpricing.replace({"Part Pricing - Raymond": "Part_Pricing", "Part Pricing - TVH/Durasource":"Part_Pricing",
                                    "Part Pricing - Misc":"Part_Pricing", "Tax":"Tax_Enviro", "Environmental Fee":"Tax_Enviro", "Over NTE":"Over_NTE", 
                                    "PO Funds":"Over_NTE", "PO Funds":"Over_NTE", "Incorrect Labor Rate":"Incorrect_Rate",
                                    "Discount Labor Line":"Incorrect_Rate", "Incorrect PM Rate":"Incorrect_Rate", "Incorrect Part Pricing":
                                    "Incorrect_Rate", "Incorrect Labor Rate/Amt":"Incorrect_Rate" , "Productive Rounding":"Incorrect_Rate",
                                    "Entry Error - Labor Quantity":"Entry_Error", "Entry Error - Freight": "Entry_Error", 
                                    "Entry Error - Travel Quantity":"Entry_Error", "Entry Error - Misc Charge":"Entry_Error", 
                                    "Entry Error - Part Pricing":"Entry_Error", "Entry Error - Travel Rate":"Entry_Error", "Entry Error - Part Quantity":
                                    "Entry_Error", "Entry Error - SM Pricing":"Entry_Error", "PDN":"Misc_Charge", "Tire Press":"Misc_Charge", 
                                    "Tire Disposal":"Misc_Charge", "Entry Error - Tire Press":"Misc_Charge", "Incorrect Travel Rate":"Incorrect_Rate",
                                    "Incorrect Vendor Markup":"Incorrect_Rate", "Incorrect Battery Rate":"Incorrect_Rate", "Flat Rate Error":"Incorrect_Rate",
                                    "Incorrect Travel Quantity":"Incorrect_Rate", "Incorrect Travel Rate":"Incorrect_Rate" , "Unpaid - Reversed Status":"Unpaid", "Unpaid - Cancel Status":"Unpaid",
                                    "Unpaid - Denied Status":"Unpaid", "Unpaid - Return Status":"Unpaid", "Entry Error - Duplicate Claim":"Entry_Error", "Core Charge":"Entry_Error",
                                    "Entry Error - Labor Dates":"Entry_Error" }) 

invpricing["Cause"].value_counts()
```

```python
#create new DF for each Branch/Dept to calculate monthly impact, $$ per cause, $$ per customer
denService_df = invpricing[invpricing.Branch_Dept == "DEN_Service"]
slcService_df = invpricing[invpricing.Branch_Dept == "SLC_Service"]
abqService_df = invpricing[invpricing.Branch_Dept == "ABQ_Service"]
denShop_df = invpricing[invpricing.Branch_Dept == "DEN_Shop"]
denParts_df = invpricing[invpricing.Branch_Dept == "DEN_Parts"]
slcParts_df = invpricing[invpricing.Branch_Dept == "SLC_Parts"]
abqParts_df = invpricing[invpricing.Branch_Dept == "ABQ_Parts"]
denTire_df = invpricing[invpricing.Branch_Dept == "DEN_Tire"]
```

```python
#calculate monthly discrepancies for each Branch/Dept

#DEN Service
DenService_monthly = denService_df.groupby(["Month"]).sum("Discrepancy")
DenService_monthly = DenService_monthly["Discrepancy"]


#SLC Service
SlcService_monthly = slcService_df.groupby(["Month"]).sum("Discrepancy")
SlcService_monthly = SlcService_monthly["Discrepancy"]

#ABQ Service
AbqService_monthly = abqService_df.groupby(["Month"]).sum("Discrepancy")
AbqService_monthly = AbqService_monthly["Discrepancy"]

#DEN Shop
DenShop_monthly = denShop_df.groupby(["Month"]).sum("Discrepancy")
DenShop_monthly = DenShop_monthly["Discrepancy"]

#DEN Parts
DenParts_monthly = denParts_df.groupby(["Month"]).sum("Discrepancy")
DenParts_monthly = DenParts_monthly["Discrepancy"]

#SLC Parts
SlcParts_monthly = slcParts_df.groupby(["Month"]).sum("Discrepancy")
SlcParts_monthly = SlcParts_monthly["Discrepancy"]

#ABQ Parts
AbqParts_monthly = abqParts_df.groupby(["Month"]).sum("Discrepancy")
AbqParts_monthly = AbqParts_monthly["Discrepancy"]

#DEN Tire
DenTire_monthly = denTire_df.groupby(["Month"]).sum("Discrepancy")
DenTire_monthly = DenTire_monthly["Discrepancy"]
```
